```python
import matplotlib.pyplot as plt
import numpy as np

def pressure(maxMassFlowRate, numberOfCylinders):
    iManPressureSwitch = 805 #* u.psi
    iCylinderBasis = 1
    CylFlow = maxMassFlowRate / iCylinderBasis
    manifoldPressure = -119.0 * CylFlow ** 3.0 + 251.21 * CylFlow**2.0-152.13*CylFlow+1162.3
    while(iCylinderBasis<numberOfCylinders and manifoldPressure<iManPressureSwitch and numberOfCylinders>1):
        iCylinderBasis = iCylinderBasis +1
        CylFlow = maxMassFlowRate/iCylinderBasis
        manifoldPressure = -119.0 * CylFlow ** 3.0 + 251.21 * CylFlow**2.0-152.13*CylFlow+1162.3
    return manifoldPressure
```

```python
n = 1001
nc = 6
massFlowRate = np.linspace(0, 15,n)

p = []
for i in (massFlowRate):
    p.append(pressure(i,nc))

plt.plot(massFlowRate, p, label="P vs. $\dot{m}$")
plt.xlabel('$\dot{m}$ (lb/s)')
plt.ylabel('P (psi)')
plt.grid()
plt.legend()
plt.show()
```
